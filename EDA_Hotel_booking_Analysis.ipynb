# **Checking if data set contain any Null values**

From above code Statement came to know that maximum Null values are present at the columns agent and company & Some null values in children and Country column. we will drop company column as its has only 5.7% valus as non Null. We will fill null values of other column i.e. agent, Children, Country with feature Engineering

```python
df.drop(columns=["company"],inplace=True)
```

Verifying that drop has done successfully

```python
df.columns
```

Agent Column has around 15.39 % data as Null value and we can drop a column which has around 13.69 % data, so we will enrich/fill with '0'. Country Cloumn as there are total 195 contarys in the world. we will replace null values of country column to 'other'.

```python
df['agent'].fillna(0,inplace=True)
mode_of_country = df['country'].mode()
df['country'].fillna('mode_of_country',inplace=True)
```
